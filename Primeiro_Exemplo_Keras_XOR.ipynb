```python
import numpy as np
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])
```

```python
rede = Sequential()
```

```python
rede.add(Dense(units = 3, activation ='sigmoid', input_dim=2))
rede.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
rede.compile(loss = 'mean_squared_error')
```

```python
rede.fit(X, y, epochs = 10000)
```

```python
rede.evaluate(X, y)
```

```python
rede.predict(X)
```
