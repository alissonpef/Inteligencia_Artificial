```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/Credit Score Classification Dataset.csv')
```

```python
dados.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb_encoder = LabelEncoder()
```

```python
dados['Gender'] = lb_encoder.fit_transform(dados['Gender'])
dados['Education'] = lb_encoder.fit_transform(dados['Education'])
dados['Marital Status'] = lb_encoder.fit_transform(dados['Marital Status'])
dados['Home Ownership'] = lb_encoder.fit_transform(dados['Home Ownership'])
dados['Credit Score'] = lb_encoder.fit_transform(dados['Credit Score'])
```

```python
dados.head()
```

```python
dados.shape
```

```python
X = dados.iloc[:, 0:7].values
y = dados.iloc[:, 7].values
```

```python
X, y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, train_size=0.7)
```

# Teste do KNN e Árvore de Decisão

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
arvore = DecisionTreeClassifier(criterion="entropy")
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
arvore.fit(X_treino, y_treino)
knn.fit(X_treino, y_treino)
```

```python
print(f"Score Arvore de Decisao: {arvore.score(X_teste, y_teste)}")
print(f"Score KNN: {knn.score(X_teste, y_teste)}")
```
